```python
import pandas as pd
df=pd.read_csv("/content/marksheet.csv")
print(df)
```

## ***Total Student***

```python
total_students = len(df)
print(total_students)
```

# ***Average marks***

```python
average_marks = df['Total'].mean()
print(average_marks)
```

# ***Highest / Lowest Marks***

```python
highest_score = df['Total'].max()
print(highest_score)
```

```python

lowest_score = df['Total'].min()
print(lowest_score)
```

# ***Topper of the class***

```python
topper = df.loc[df['Total'].idxmax()]
print(topper)
```
